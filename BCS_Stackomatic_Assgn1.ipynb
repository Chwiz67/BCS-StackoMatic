# Part 1: Data Exploration with Pandas

Load the Dataset

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/california_housing_train.csv')
```

```python
df.head()
```

```python
df.tail(10)
```

Data Summary

```python
df.describe()
```

```python
print(f'Unique values in total_bedrooms: {df.total_bedrooms.unique()}')
print(f'No. of unique values in total_bedrooms: {df.total_bedrooms.nunique()}')
```

Data Transformation

```python
df['total_bedrooms_per_total_rooms'] = df['total_bedrooms'] / df['total_rooms']
mean = df['total_bedrooms_per_total_rooms'].mean()
std = df['total_bedrooms_per_total_rooms'].std()
print(f'Mean of total_bedrooms_per_total_rooms = {mean}')
print(f'Standard Deviation of total_bedrooms_per_total_rooms = {std}')
```

Data Filtering

```python
df[df['median_income']>5]
```

```python
df[(df['total_rooms']>10000) & (df['median_house_value']<150000)]
```

Export Modified Data

```python
df.to_csv('/content/sample_data/modified_california_housing_train.csv', index=False)
```

#Part 2: Numerical Analysis with NumPy

Array Creation and Indexing

```python
import numpy as np
```

```python
arr1 = np.arange(1, 21)
arr2 = np.random.choice(range(10,101), size=(5,4))
print(arr1)
print(arr1[1::2])
print(arr2)
print(arr2[:3, :2])
```

Array Manipulation

```python
arr1 = arr1.reshape(4,5)
arr2 = arr2.flatten()
arr3 = np.random.randint(1, 21, size=(5,4))
print(arr2)
print(np.dot(arr1, arr3))
```

Boolean and Fancy Indexing

```python
arr4 = np.random.randint(1, 101, size=(50))
idx = np.where(arr4>50)
print(arr4[idx])
print(arr4[[3,6,23,46,49]])
```

Statistical Operations

```python
print(arr1.mean(axis=0))
print(arr1.sum(axis=0))
print(arr1.max(axis=0))
```

```python
mask = np.array([arr1[:,i]>arr1.mean(axis=0)[i] for i in range(5)]).T
print(arr1[mask])
```

#Part 3: Problem Solving with Python Basics

Loops and Conditional Statements

```python
fib = [0, 1]
for i in range(2, 100):
  fib.append(fib[i-1] + fib[i-2])
print(fib)
```

```python
for i in range(1, 51):
  if i%3==0 and i%5==0:
    print(f'{i}: FizzBuzz')
  elif i%3==0:
    print(f'{i}: Fizz')
  elif i%5==0:
    print(f'{i}: Buzz')
```

Variable-Length Arguments

```python
n1 = input('Enter <space> separated nos. (Eg: 2 3 34 45): ').split()
n1 = [int(i) for i in n1]
n1.sort()
if len(n1)%2==0:
  print(f'Median = {(n1[len(n1)//2] + n1[len(n1)//2 - 1]) / 2}')
else:
  print(f'Median = {n1[len(n1)//2]}')
```

```python
n2 = input('Enter <space> separated nos. (Eg: 2 3 34 45): ').split()
n2 = [int(i) for i in n2]
print(f'Mean = {sum(n2) / len(n2)}')
```

Error Handling

```python
try:
  1/0
except ZeroDivisionError:
  print('Cannot divide by zero')
```
